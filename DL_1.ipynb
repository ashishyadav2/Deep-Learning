## Deep Learning ANN Implementation

### Sequential: Forward Propagation
### Dense: Neurons, nodes in ANN
### Activation Function: LeakyReLU,  PReLU,  ELU,  ReLU
### Dropout: To avoid overfitting of data it will drop few percentage of neurons


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, PReLU, ELU, ReLU, Dropout
```

```python
# Initialize ANN
classifier = Sequential()
```

```python
# Adding Input layer
# .add(neuron nodes (Dense(units = number of nodes))) units will be always be equal to number of independent variables
# activation function will apply to next layer after where it is defined.
# here activation function is defined in input layer but it will be applied to hidden layer
# and activation fucntion defined in hidden layer will apply in output layer or hidden layer depending on the layers used
classifier.add(Dense(units=5,activation=ReLU))
```

```python
## Addiing Hidden Layer
## drop 30% of neurons/ nodes
classifier.add(Dense(units=7,activation=ReLU))
classifier.add(Dropout(0.3))
```

```python
## Adding second hidden layer
classifier.add(Dense(units=5,activation=ReLU))
```

```python
## Adding Output Layer
## in binary classifier sigmoid is used always it give 0 or 1 as output
## val<0.5 returns 0, val >= 0.5 returns 1
## in multiclass classification softmax is used in output layer
classifier.add(Dense(1,activation='sigmoid'))
```
